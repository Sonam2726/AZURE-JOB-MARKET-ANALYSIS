```python
from pyspark.sql.functions import *
```

```python
spark.conf.set(
    f"fs.azure.account.key.<storagename>.dfs.core.windows.net",
    <"storagekey">
)
```

```python
source_csv_path = "abfss://data@projectstorage2612.dfs.core.windows.net/jobs.csv"

bronze_path = "abfss://bronze@projectstorage2612.dfs.core.windows.net/"
```

```python
#read data
df = spark.read.format("csv") \
.option("header", "true") \
.option("inferSchema", "true") \
.option("multiLine", "true") \
.option("quote", "\"") \
.option("escape", "\"") \
.load(source_csv_path)
```

```python
jobs_df = df.select(
    "job_id",
    "title",
    "description",
    "company_id"
)

jobs_df.write.format("delta").mode("overwrite").save(bronze_path + "jobs_raw")


```

```python
company_df = df.select(
    "company_id",
    "company_name"
).dropDuplicates()

company_df.write.format("delta").mode("overwrite").save(bronze_path + "company_raw")
```

```python
salary_df = df.select(
    "job_id",
    "min_salary",
    "med_salary",
    "max_salary",
    "normalized_salary",
    "currency",
    "compensation_type",
    "pay_period"
)

salary_df.write.format("delta").mode("overwrite").save(bronze_path + "salary_raw")
```

```python
location_df = df.select(
    "job_id",
    "location",
    "zip_code",
    "fips",
    "remote_allowed"
)

location_df.write.format("delta").mode("overwrite").save(bronze_path + "location_raw")
```

```python
experience_df = df.select(
    "job_id",
    "formatted_experience_level",
    "formatted_work_type",
    "work_type"
)

experience_df.write.format("delta").mode("overwrite").save(bronze_path + "experience_raw")
```

```python
posting_df = df.select(
    "job_id",
    "original_listed_time",
    "listed_time",
    "expiry",
    "closed_time",
    "posting_domain",
    "sponsored",
    "applies",
    "views",
    "application_type"
)

posting_df.write.format("delta").mode("overwrite").save(bronze_path + "posting_raw")
```
